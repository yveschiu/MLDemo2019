```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
df = pd.DataFrame(data=boston["data"], columns=boston["feature_names"])
df
```

```python
df["target"] = boston["target"]
```

```python
df
```

```python
# heatmap
plt.figure(figsize=(14,10))
plt.title("Pearson Correlation of Features", y=1.05, size=15)
sns.heatmap(df.astype(float).corr(), linewidths=0.1, square=True, linecolor="white", annot=True)
```

```python
# è¨“ç·´æ¨¡åž‹
# å°‡è³‡æ–™åˆ‡æˆ è¨“ç·´/æ¸¬è©¦ 9/1
from sklearn.model_selection import train_test_split
data_train, data_test, target_train, target_test = train_test_split(boston["data"], boston["target"], test_size=0.1) 
```

```python
from sklearn.tree import DecisionTreeRegressor
regr = DecisionTreeRegressor(max_depth=3)
regr.fit(data_train,target_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz


dot_data = export_graphviz(regr, 
                           out_file=None,
                           feature_names=boston["feature_names"], 
                           filled=True, 
                           rounded=True, 
                           special_characters=True)

graph = graphviz.Source(dot_data)
# è¼¸å‡º pdf
# graph.render("iris2")
graph
```

```python
# numpy å¯ä»¥è®“å…©å€‹ list ç›´æŽ¥ç›¸æ¸›
import numpy as np
predict = regr.predict(data_test)
print("å¯¦éš›çš„åƒ¹éŒ¢ï¼š", target_test)
print("é æ¸¬çš„åƒ¹éŒ¢ï¼š", predict)
interval = np.subtract(predict, target_test)
print("å·®ç•°ï¼š", interval)
```
