```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
```

```python
import pandas as pd
iris = load_iris()
df = pd.DataFrame(iris["data"], columns=iris["feature_names"])
df["target"] = iris["target"]
df
```

```python
# 切開來 --> 為之後步驟做準備
from sklearn.model_selection import train_test_split
# [90%題目, 10%題目, 90%答案, 10%答案]
x_train, x_test, y_train, y_test = train_test_split(iris["data"], iris["target"], test_size=0.1)
```

```python
# clu (cluster)
clu = KMeans(n_clusters=3)
clu.fit(x_train)
```

```python
print("預測:", list(clu.predict(x_test)))
```

```python
print("本來的標籤", list(y_test))
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(clu.predict(x_test), y_test)
```

```python
# 拿出資料點被打上的標籤
clu.labels_
```

```python
from sklearn.metrics import silhouette_score
xs = []
ys = []
for k in range(2,6):
    clu = KMeans(n_clusters=k)
    clu.fit(iris["data"])
    xs.append(k)
    score = silhouette_score(iris["data"], clu.labels_)
    ys.append(score)
print(xs)
print(ys)
    
```

```python
import matplotlib.pyplot as plt
#省略每次plt.plot()後要打plt.show()，寫了這行之後都不用再寫.%是jupyter特有符號，原生python沒有
%matplotlib inline
plt.title("Silhouette Score")
plt.xlabel("K")
plt.ylabel("Score")
plt.plot(xs, ys, "m^--")
# plt.show() 
```
