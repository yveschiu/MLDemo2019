```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(data = iris["data"], columns=iris["feature_names"])
df
```

```python
df["target"] = iris["target"]
```

```python
df
```

```python
# heatmap --> 觀察相關係數
# 花瓣的長度與寬度與判斷品種幾乎呈現完全正相關
plt.figure(figsize=(14,10)) # create a new figure. figsize = (width, height) in inches
plt.title("Pearson Correlation of Features", y=1.05, size=15)
sns.heatmap(df.astype(float).corr(),
            cmap="BrBG", 
            linewidths=0.1, 
            square=True,
            linecolor="white", 
            annot=True)
```

```python
# 將資料及分成兩份，一份訓練集和一份測試集
from sklearn.model_selection import train_test_split
# 切成 9/1 
data_train, data_test, target_train, target_test = train_test_split(iris["data"], iris["target"], test_size=0.1)
```

```python
pd.DataFrame(data_train, columns=iris["feature_names"]).head()
```

```python
# 分類問題 --> 使用 DecisionTreeClassifier 來訓練模型
from sklearn.tree import DecisionTreeClassifier

# 生成一棵最大深度為3的數，並使用 clf.fit() 方法餵資料來訓練模型
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(data_train, target_train)
```

```python
# 使用第三方軟體 graphiz 來繪圖
# 載入 python 的 graphiz library
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(clf, 
                           out_file=None,
                           feature_names=iris.feature_names,
                           class_names=iris.target_names, 
                           filled=True, 
                           rounded=True, 
                           special_characters=True)

graph = graphviz.Source(dot_data)
# 輸出 pdf
# graph.render("iris2")
graph
```

```python
# 用訓練好的模型來測試沒看過的資料 clf.predict()方法
# 使用 accuracy_score 來算正確率
from sklearn.metrics import accuracy_score

predict = clf.predict(data_test) # 只要餵測試資料就好，回傳預測答案的array
print("預測：", predict)
print("正確答案：", target_test)
print("正確率：", accuracy_score(target_test, predict) * 100, "%")
```

```python
# 使用混淆矩陣看分類錯誤
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target_test, predict)
# 用 pandas 包起來比較好判讀 對角線為分類正確的結果
pd.DataFrame(cm, columns=iris.target_names)
```

```python
# 看決策樹的變數重要性
# 可以和前面畫的熱度圖（相關係數）比較，看看重要程度的次序是否一樣
print(iris.feature_names)
print(clf.feature_importances_)
```
