```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(data = iris["data"], columns=iris["feature_names"])
df
```

```python
df["target"] = iris["target"]
```

```python
df
```

```python
# heatmap --> è§€å¯Ÿç›¸é—œä¿‚æ•¸
# èŠ±ç“£çš„é•·åº¦èˆ‡å¯¬åº¦èˆ‡åˆ¤æ–·å“ç¨®å¹¾ä¹Žå‘ˆç¾å®Œå…¨æ­£ç›¸é—œ
plt.figure(figsize=(14,10)) # create a new figure. figsize = (width, height) in inches
plt.title("Pearson Correlation of Features", y=1.05, size=15)
sns.heatmap(df.astype(float).corr(),
            cmap="BrBG", 
            linewidths=0.1, 
            square=True,
            linecolor="white", 
            annot=True)
```

```python
# å°‡è³‡æ–™åŠåˆ†æˆå…©ä»½ï¼Œä¸€ä»½è¨“ç·´é›†å’Œä¸€ä»½æ¸¬è©¦é›†
from sklearn.model_selection import train_test_split
# åˆ‡æˆ 9/1 
data_train, data_test, target_train, target_test = train_test_split(iris["data"], iris["target"], test_size=0.1)
```

```python
pd.DataFrame(data_train, columns=iris["feature_names"]).head()
```

```python
# åˆ†é¡žå•é¡Œ --> ä½¿ç”¨ DecisionTreeClassifier ä¾†è¨“ç·´æ¨¡åž‹
from sklearn.tree import DecisionTreeClassifier

# ç”Ÿæˆä¸€æ£µæœ€å¤§æ·±åº¦ç‚º3çš„æ•¸ï¼Œä¸¦ä½¿ç”¨ clf.fit() æ–¹æ³•é¤µè³‡æ–™ä¾†è¨“ç·´æ¨¡åž‹
clf = DecisionTreeClassifier(max_depth=3)
clf.fit(data_train, target_train)
```

```python
# ä½¿ç”¨ç¬¬ä¸‰æ–¹è»Ÿé«” graphiz ä¾†ç¹ªåœ–
# è¼‰å…¥ python çš„ graphiz library
from sklearn.tree import export_graphviz
import graphviz

dot_data = export_graphviz(clf, 
                           out_file=None,
                           feature_names=iris.feature_names,
                           class_names=iris.target_names, 
                           filled=True, 
                           rounded=True, 
                           special_characters=True)

graph = graphviz.Source(dot_data)
# è¼¸å‡º pdf
# graph.render("iris2")
graph
```

```python
# ç”¨è¨“ç·´å¥½çš„æ¨¡åž‹ä¾†æ¸¬è©¦æ²’çœ‹éŽçš„è³‡æ–™ clf.predict()æ–¹æ³•
# ä½¿ç”¨ accuracy_score ä¾†ç®—æ­£ç¢ºçŽ‡
from sklearn.metrics import accuracy_score

predict = clf.predict(data_test) # åªè¦é¤µæ¸¬è©¦è³‡æ–™å°±å¥½ï¼Œå›žå‚³é æ¸¬ç­”æ¡ˆçš„array
print("é æ¸¬ï¼š", predict)
print("æ­£ç¢ºç­”æ¡ˆï¼š", target_test)
print("æ­£ç¢ºçŽ‡ï¼š", accuracy_score(target_test, predict) * 100, "%")
```

```python
# ä½¿ç”¨æ··æ·†çŸ©é™£çœ‹åˆ†é¡žéŒ¯èª¤
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(target_test, predict)
# ç”¨ pandas åŒ…èµ·ä¾†æ¯”è¼ƒå¥½åˆ¤è®€ å°è§’ç·šç‚ºåˆ†é¡žæ­£ç¢ºçš„çµæžœ
pd.DataFrame(cm, columns=iris.target_names)
```
