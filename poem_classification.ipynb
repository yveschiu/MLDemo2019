```python
import pandas as pd
traindf = pd.read_csv("poem_train.csv", encoding="utf-8")
traindf
```

```python
testdf = pd.read_csv("poem_test.csv", encoding="utf-8")
testdf
```

```python
import jieba
# 實驗結果沒有load更大的分詞字典比較準
# jieba.load_userdict("dict.txt.big")
```

```python
s = "平林漠漠煙如織，寒山一帶傷心碧。\r\n暝色入高樓，有人樓上愁。".replace("\r", "").replace("\n", "")
" ".join(jieba.cut(s))
```

```python
def poemflow(p):
    s = p.replace("\r", "").replace("\n", "")
    return " ".join(jieba.cut(s))
traindf["內容"] = traindf["內容"].apply(poemflow)
traindf
```

```python
testdf["內容"] = testdf["內容"].apply(poemflow)
testdf
```

```python
# 分類的時候，答案一定要轉成 0, 1, 2, 3, 4
trans = {"李白":0, "杜甫":1, "白居易":2}
traindf["作者"] = traindf["作者"].replace(trans)
traindf
```

```python
testdf["作者"] = testdf["作者"] = testdf["作者"].replace(trans)
testdf
```

```python
# 詩名我不要 drop
# 對dataframe做操作要指定axis， axis=0 --> 列操作  axis=1 --> 行操作
traindf = traindf.drop(["詩名"], axis=1)
traindf
```

```python
testdf = testdf.drop(["詩名"], axis=1)
testdf
```

```python
# 只有訓練才fit
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
train_bag = vec.fit_transform (traindf["內容"])
# 輸出
# <2731x47259 sparse matrix of type '<class 'numpy.int64'>'
# 	with 88353 stored elements in Compressed Sparse Row format>

test_bag = vec.transform(testdf["內容"])
```

```python
vec.vocabulary_
```

```python
print(train_bag)
```

```python
# 機器學習開始
# 做出公式
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB()
clf.fit(train_bag, traindf["作者"])
# 輸出 MultinomialNB(alpha=1.0, class_prior=None, fit_prior=True)
# alpha = 1 --> Laplace smoothing
```

```python
# # 驗證一下公式好壞
pre = clf.predict(test_bag)
print("預測： ", list(pre))
print("答案：", list(testdf["作者"]))
```

```python
from sklearn.metrics import accuracy_score
print("正確率：", accuracy_score(pre, testdf["作者"]))
```

```python
# 反字典
reversetrans = {trans[key]:key for key in trans}
reversetrans
```

```python
s = input("隨便輸入一首詩：")
s = " ".join(jieba.cut(s))
sbag = vec.transform([s])
print(clf.predict(sbag))
print("答案：", reversetrans[clf.predict(sbag)[0]])
```
